```python
import matplotlib.pyplot as plt
import jax.numpy as jnp
import seaborn as sns
import pickle
```

```python
# sample size: 108
with open('mala_chains_108_ideal_sensors.pkl', 'rb') as f:
    mala_chains_108 = pickle.load(f)
```

```python
# sample size: 1_000
with open('mala_chains_1000_ideal_sensors.pkl', 'rb') as f:
    mala_chains_1000 = pickle.load(f)
```

```python
# sample size: 5_000
with open('mala_chains_5000_ideal_sensors.pkl', 'rb') as f:
    mala_chains_5000 = pickle.load(f)
```

```python
burn_in = 0
```

## 1. True Emission Rate

```python
sns.kdeplot(mala_chains_108["s"][burn_in:,5], color='yellow', label='108', fill=True, alpha=0.1)
sns.kdeplot(mala_chains_1000["s"][burn_in:,5], color='blue', label='1_000', fill=True, alpha=0.1)
sns.kdeplot(mala_chains_5000["s"][burn_in:,5], color='seagreen', label='5_000', fill=True, alpha=0.1)
plt.axvline(1.6e-3, color='red', linestyle='--')
plt.xlabel("Emission rate")
plt.ylabel("Density")
plt.xlim(0.001,0.002)
plt.title("Source emission rate")
plt.legend()
plt.show()
```

## 2. Zero emission cells

```python
jnp.delete(mala_chains_108["s"][burn_in:], 5, axis=1)
sns.kdeplot(jnp.sum(jnp.delete(mala_chains_108["s"][burn_in:], 5, axis=1), axis=1), color='yellow', label='108', fill=True, alpha=0.1)
sns.kdeplot(jnp.sum(jnp.delete(mala_chains_1000["s"][burn_in:], 5, axis=1), axis=1), color='blue', label='1_000', fill=True, alpha=0.1)
sns.kdeplot(jnp.sum(jnp.delete(mala_chains_5000["s"][burn_in:], 5, axis=1), axis=1), color='seagreen', label='5_000', fill=True, alpha=0.1)
plt.axvline(0.0, color='red', linestyle='--')
plt.xlabel("Sum of emission rates")
plt.ylabel("Density")
plt.xlim(-1e-4,0.001)
plt.title("Sum of all cells without emissions")
plt.legend()
plt.show()
```

## 2. Sigma Squared

```python
sns.kdeplot(mala_chains_108["sigma_squared"][burn_in:].squeeze(), color='yellow', label='108', fill=True, alpha=0.1)
sns.kdeplot(mala_chains_1000["sigma_squared"][burn_in:].squeeze(), color='blue', label='1_000', fill=True, alpha=0.1)
sns.kdeplot(mala_chains_5000["sigma_squared"][burn_in:].squeeze(), color='seagreen', label='5_000', fill=True, alpha=0.1)
plt.axvline(1e-3, color='red', linestyle='--')
plt.xlabel("Sigma squared")
plt.ylabel("Density")
plt.title("Source emission rate")
plt.legend()
plt.show()
```

## Horizontal Wind Sigmas

```python
# Create a figure and two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot data on the first subplot
sns.kdeplot(mala_chains_108["tan_gamma_H"][burn_in:].squeeze(), color='yellow', label='108', fill=True, alpha=0.1, ax=ax1)
sns.kdeplot(mala_chains_1000["tan_gamma_H"][burn_in:].squeeze(), color='blue', label='1000', fill=True, alpha=0.1, ax=ax1)
sns.kdeplot(mala_chains_5000["tan_gamma_H"][burn_in:].squeeze(), color='seagreen', label='5000', fill=True, alpha=0.1, ax=ax1)
ax1.axvline(0.36397022, color='red', linestyle='--')
ax1.set_xlabel("tan_gamma_H")
ax1.set_ylabel("Density")
ax1.set_title('First Density Plot')
ax1.legend()

# Plot data on the second subplot
sns.kdeplot(mala_chains_108["b_H"][burn_in:].squeeze(), color='yellow', label='108', fill=True, alpha=0.1, ax=ax2)
sns.kdeplot(mala_chains_1000["b_H"][burn_in:].squeeze(), color='blue', label='1000', fill=True, alpha=0.1, ax=ax2)
sns.kdeplot(mala_chains_5000["b_H"][burn_in:].squeeze(), color='seagreen', label='5000', fill=True, alpha=0.1, ax=ax2)
ax2.axvline(1.0, color='red', linestyle='--')
ax2.set_xlabel("b_H")
ax2.set_ylabel("Density")
ax2.set_title('Second Density Plot')
ax2.legend()

# Display the figure with two subplots
plt.show()
```
